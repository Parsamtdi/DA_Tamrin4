```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

data = pd.read_csv('C:\\Users\\admin\\Desktop\\mojtamafani\\tamrin4\\sample.csv')  
LokLat = data['LocationLat'].values 

# Create the plot
plt.scatter(np.arange(len(LokLat)) + 1, LokLat, label='Data Points')

# Calculate statistics
mean = LokLat.mean()
std = LokLat.std()
upper_bond = mean + std * 2
lower_bond = mean - std * 2

# khotot miangin va lower upper bond
plt.plot([0,len(LokLat)],[mean,mean],"g--",label="Mean")
plt.plot([0,len(LokLat)],[upper_bond,upper_bond],"r--",label="Upper_bond")
plt.plot([0,len(LokLat)],[lower_bond,lower_bond],"b--",label="Lower_bond")

# ezafe kardan labels and legend
plt.xlabel('Data Point Index')
plt.ylabel('LokLLat')
plt.title('LocationLat Data with Statistical Boundaries')
plt.legend()

plt.show()

# mohasebe va print outliers
outliers = LokLat[(LokLat > upper_bond) | (LokLat < lower_bond)]
print("Outliers:", outliers)
```
